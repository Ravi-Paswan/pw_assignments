# Question 01

Matplotlib is a Python library used for data visualization. It allows users to create a wide range of high-quality 2D and 3D plots, including line plots, scatter plots, bar plots, histograms, and many others.

Matplotlib is used to create graphs and visualizations for scientific research, engineering, finance, and other fields that require the analysis and presentation of large amounts of data.

The Pyplot module is a sublibrary of Matplotlib that provides a simple interface for creating plots. Some of the commonly used plots that can be created using the Pyplot module include:

Line plot: A plot that displays data as a series of points connected by straight lines.

Scatter plot: A plot that displays data as a collection of points, with each point representing a value for two variables.

Bar plot: A plot that displays data as a series of rectangular bars, with the height of each bar representing the value of a variable.

Histogram: A plot that displays the distribution of a dataset by dividing it into a set of intervals and counting the number of values that fall into each interval.

Box plot: A plot that displays the distribution of a dataset by showing the range, median, and quartiles of the data.

# Question 02

Scatter Plot - A scatter plot is a type of plot used in data visualization that displays the values of two variables as a set of points on a two-dimensional coordinate system. The horizontal axis of the plot represents one variable, and the vertical axis represents the other variable.

Each point on the plot represents a pair of values for the two variables being plotted. Scatter plots are useful for identifying patterns and relationships between two variables. They can help to identify correlations, outliers, clusters, and other patterns in the data.

For example, a scatter plot can be used to display the relationship between the height and weight of a group of people. Each point on the plot would represent the height and weight of a single person, and the scatter plot could be used to identify any patterns or relationships between the two variables.

Scatter plots can also be used to visualize datasets with more than two variables by using color, size, or shape to represent additional variables. In these cases, the plot becomes a 3D or higher-dimensional scatter plot, but the basic idea remains the same: to display the relationship between multiple variables by representing them as points on a plot.

```python
np.random.seed(3)
x = 3+ np.random.normal(0,2,50)
y = 3+ np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

# Question 03

The subplot function is used in Matplotlib to create multiple plots in a single figure. It allows users to arrange multiple plots in a grid or other custom layout, making it easier to compare and visualize different aspects of the data.

There are several reasons why subplot is a useful function:

Compare multiple plots side by side: By using subplot, multiple plots can be arranged side by side in a single figure, allowing users to compare them easily. For example, you might want to compare different visualizations of the same dataset, such as a scatter plot and a histogram.

Save space: Using subplot can save space when creating multiple plots. Instead of creating multiple figures, all the plots can be arranged in a single figure, reducing the amount of screen real estate required to view the plots.

Customization: subplot provides flexibility in terms of customizing the layout of the plots. Users can specify the number of rows and columns in the grid, and can also adjust the spacing between the plots.

Better visualization: Creating multiple plots using subplot can improve the visualization of the data by allowing users to explore different aspects of the data in a single figure. This can make it easier to identify patterns and trends in the data.

Overall, subplot is a powerful tool for creating multiple plots in Matplotlib, providing flexibility, customization, and improved visualization of the data.





```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure object
fig = plt.figure()

# Create a grid of subplots
ax1 = fig.add_subplot(2, 2, 1) # Upper left
ax2 = fig.add_subplot(2, 2, 2) # Upper right
ax3 = fig.add_subplot(2, 2, 3) # Lower left
ax4 = fig.add_subplot(2, 2, 4) # Lower right

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Plot the data in each subplot
ax1.plot(np.array([0,1,2,3,4,5]),np.array([0, 100, 200, 300, 400, 500]))
ax2.plot(np.array([0,1,2,3,4,5]),np.array([50, 20, 40, 20, 60, 70]))
ax3.plot(np.array([0,1,2,3,4,5]),np.array([10, 20, 30, 40, 50, 60]))
ax4.plot(np.array([0,1,2,3,4,5]),np.array([200, 350, 250, 550, 450, 150]))


# Add titles to each subplot
ax1.set_title('Line 1')
ax2.set_title('Line 2')
ax3.set_title('Line 3')
ax4.set_title('Line 4')

# Add a title to the entire figure
fig.suptitle('Four subplots')
fig.subplots_adjust(hspace=0.5, wspace=0.5)
# Show the plot
plt.show()
```

# Question 05

A box plot, also known as a box-and-whisker plot, is a type of plot in data visualization that displays the distribution of numerical data through five summary statistics: the minimum, the maximum, the median, the first quartile (25th percentile), and the third quartile (75th percentile).

The box in a box plot represents the interquartile range (IQR), which is the range between the first and third quartiles. The median is represented by a vertical line inside the box. The whiskers, which extend from the top and bottom of the box, represent the minimum and maximum values of the data that fall within 1.5 times the IQR from the first and third quartiles, respectively. Any data points outside of this range are considered outliers and are represented by dots or asterisks.

Box plots are useful for comparing the distribution of data across different groups or categories and identifying any outliers or extreme values. They can also provide a visual representation of the skewness or symmetry of the data.

In Matplotlib's pyplot module, the boxplot() function can be used to create box plots in Python. The function accepts an array or a sequence of arrays as input, where each array represents a different group or category of data to be plotted. Additional options can be specified to customize the appearance of the box plot, such as the color, the orientation, and the whisker length

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
```
