### What is Exception in python? Write the difference between Exception and Syntax error?

In Python, an exception is an event that occurs during the execution of a program that disrupts the normal flow of the program's instructions. When an exception occurs, the program's normal execution is halted, and the control is transferred to a special code block known as an exception handler. Exception handling allows you to catch and handle these exceptional events gracefully, preventing the program from crashing or producing unexpected results.

Exceptions can occur for various reasons, such as trying to access a non-existent variable, dividing a number by zero, or attempting to open a file that doesn't exist. Python provides a mechanism to handle different types of exceptions using try-except blocks. The code inside the try block is monitored for exceptions, and if any exception occurs, it is caught by the corresponding except block, where you can handle the exception appropriately.

Syntax errors, on the other hand, are a type of error that occurs when the Python interpreter encounters code that violates the language's syntax rules. These errors typically happen during the parsing phase, before the program is executed. Syntax errors indicate that there is a mistake in the structure or formatting of the code, such as missing parentheses, incorrect indentation, or using an invalid keyword.

The main difference between exceptions and syntax errors is as follows:

1. Timing: Syntax errors are caught by the Python interpreter during the parsing phase, before the program is executed. Exceptions, on the other hand, occur during the execution of the program when an exceptional condition arises.

2. Cause: Syntax errors are caused by violating the syntax rules of the Python language. Exceptions, on the other hand, are caused by exceptional situations that can occur during program execution, such as invalid input, file not found, or arithmetic errors.

3. Handling: Syntax errors cannot be caught and handled using try-except blocks because they occur before the program starts executing. However, exceptions can be caught and handled using try-except blocks to provide a graceful way of handling exceptional situations in your code.

In summary, syntax errors are mistakes in the structure or formatting of the code that prevent the program from running, while exceptions are exceptional events that occur during program execution and can be caught and handled using try-except blocks.

### What happens when an exception is not handled exception? Explain with an example

When an exception is not handled in Python, it results in the program terminating abruptly and displaying an error message called a traceback. The traceback provides information about the exception that occurred, including the type of exception, the line of code where the exception occurred, and the sequence of function calls that led to the exception.

Here's an example to illustrate what happens when an exception is not handled:

```python
def divide_numbers(a, b):
    result = a / b
    return result

num1 = 10
num2 = 0

result = divide_numbers(num1, num2)
print("Result:", result)
```

In this example, we have a function called `divide_numbers` that takes two numbers as input and divides them. We then call this function with `num1` as 10 and `num2` as 0, which will result in a ZeroDivisionError because dividing a number by zero is not allowed.

When we run this code, it will raise an exception and display a traceback:

```
Traceback (most recent call last):
  File "example.py", line 7, in <module>
    result = divide_numbers(num1, num2)
  File "example.py", line 2, in divide_numbers
    result = a / b
ZeroDivisionError: division by zero
```

As you can see from the traceback, the exception occurred in the `divide_numbers` function on line 2 when the division by zero was attempted. Since there is no exception handling code to catch and handle this exception, the program terminates immediately after the exception is raised.

Without proper exception handling, the program's flow is interrupted, and any code that was supposed to execute after the point of the exception will not run. This can lead to unexpected program behavior, crashes, or data corruption if the exception is not properly handled.

To handle exceptions and prevent the program from crashing, you can use try-except blocks to catch specific exceptions and handle them gracefully.

### Which Python statements are used to catch and handle exception? Explain with an example.

In Python, you can catch and handle exceptions using the `try-except` statement. The `try` block contains the code that you want to monitor for exceptions, and the `except` block specifies the code to be executed if an exception of a particular type is raised. Here's the syntax:

```python
try:
    # Code that might raise an exception
except ExceptionType:
    # Code to handle the exception
```

Let's consider an example to demonstrate the usage of `try-except` statement:

```python
def divide_numbers(a, b):
    try:
        result = a / b
        print("Result:", result)
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")

num1 = 10
num2 = 0

divide_numbers(num1, num2)
```

In this example, we have a function called `divide_numbers` that takes two numbers as input and attempts to divide them. Inside the `try` block, the division operation `a / b` is performed, which can potentially raise a `ZeroDivisionError` if the second number (`b`) is zero.

By enclosing the potentially problematic code within a `try` block, we can catch and handle the exception. In the `except` block, we specify the type of exception we want to handle, which in this case is `ZeroDivisionError`. If this exception occurs, the code inside the `except` block will be executed.

When we run this code, the output will be:

```
Error: Division by zero is not allowed.
```

As you can see, when the exception is raised inside the `try` block due to division by zero, the code flow is immediately transferred to the `except` block. The error message "Error: Division by zero is not allowed." is displayed instead of the program terminating abruptly.

You can catch and handle multiple types of exceptions by adding multiple `except` blocks after the `try` block, each targeting a specific exception type. Additionally, you can use a generic `except` block without specifying the exception type to catch any unexpected exceptions.

Using the `try-except` statement allows you to handle exceptions gracefully, provide appropriate error messages, and control the flow of your program even in the presence of exceptional conditions.

Certainly! I'll explain the `try-else`, `try-finally`, and `raise` statements in Python with examples.

1. `try-else` statement:
The `try-else` statement is used to specify a block of code that should be executed if no exceptions are raised within the `try` block. The `else` block is executed immediately after the `try` block completes successfully. Here's an example:

```python
try:
    # Code that might raise an exception
except ExceptionType:
    # Code to handle the exception
else:
    # Code to execute if no exceptions are raised
```

Example:

```python
try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
except ValueError:
    print("Invalid input. Please enter numbers only.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
else:
    print("Result:", result)
```

In this example, the `try` block attempts to get input from the user, convert it to integers, and perform a division operation. If a `ValueError` occurs due to non-numeric input or a `ZeroDivisionError` occurs, the corresponding `except` block is executed. If no exception is raised, the `else` block is executed, which displays the result.

2. `try-finally` statement:
The `try-finally` statement is used to define a block of code that must be executed regardless of whether an exception occurred or not. The `finally` block is executed right after the `try` block, regardless of whether an exception was raised or caught. Here's an example:

```python
try:
    # Code that might raise an exception
finally:
    # Code that will be executed regardless
```

Example:

```python
try:
    file = open("data.txt", "r")
    # Perform some operations on the file
finally:
    file.close()
    print("File closed.")
```

In this example, the `try` block attempts to open a file named "data.txt" in read mode and perform some operations on it. The `finally` block is used to ensure that the file is closed, regardless of whether an exception occurred or not.

3. `raise` statement:
The `raise` statement is used to explicitly raise an exception in Python. It allows you to throw custom exceptions or re-raise exceptions that were caught but need to be propagated further. Here's an example:

```python
try:
    # Code that might raise an exception
    if condition:
        raise ExceptionType("Error message")
except ExceptionType:
    # Code to handle the exception
```

Example:

```python
try:
    age = int(input("Enter your age: "))
    if age < 0:
        raise ValueError("Age cannot be negative.")
except ValueError as ve:
    print("Error:", str(ve))
```

In this example, the `try` block gets input for the user's age and raises a `ValueError` if the age is negative. The `except` block catches the `ValueError` and displays the error message.

The `try-else`, `try-finally`, and `raise` statements provide additional control and flexibility in handling exceptions in Python, allowing you to execute specific code paths based on whether an exception occurred or to explicitly raise exceptions when needed.

### What are Custom Exceptions in Pyhton? Why do we need Custom exceptions? Explain with example

Custom exceptions in Python are user-defined exceptions that allow you to create your own exception types specific to your application or problem domain. They are subclasses of the built-in `Exception` class or its derived classes.

We need custom exceptions for the following reasons:

1. Improved code organization: Custom exceptions help in organizing your code and making it more readable and maintainable. By defining specific exception classes, you can communicate the nature of the exceptional conditions more effectively.

2. Clearer error handling: Custom exceptions provide a way to differentiate between different types of exceptions and handle them differently. This can help in providing more informative error messages or taking specific actions based on the type of exception.

3. Modularity and reusability: Custom exceptions can be defined once and reused throughout your codebase. They encapsulate specific error conditions, making your code more modular and allowing for easier error handling across different parts of your application.

Here's an example that demonstrates the creation and usage of a custom exception:

```python
class WithdrawalError(Exception):
    """Exception raised for errors in the withdrawal process."""

    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient balance ({balance}) for withdrawal amount: {amount}")


def withdraw_amount(balance, amount):
    if amount > balance:
        raise WithdrawalError(balance, amount)
    else:
        print("Withdrawal successful.")


try:
    balance = 1000
    amount = 1500
    withdraw_amount(balance, amount)
except WithdrawalError as e:
    print("Error:", str(e))
```

In this example, we define a custom exception `WithdrawalError`, which is a subclass of `Exception`. It is raised when the withdrawal amount exceeds the available balance. The exception class takes the `balance` and `amount` as parameters and provides a custom error message using those values.

The `withdraw_amount` function checks if the withdrawal amount is greater than the balance and raises a `WithdrawalError` if it is. Otherwise, it prints a success message.

In the `try-except` block, we catch the `WithdrawalError` and print the error message. This allows us to handle the specific exception and provide a meaningful error message to the user.

Custom exceptions help in creating more expressive and understandable code, allowing you to handle exceptional situations in a precise and tailored manner. By defining your own exception classes, you can provide better error handling and communicate the exceptional conditions specific to your application.
