# Question 01

Sure, here are five functions of the Pandas library with examples of how to use them:

read_csv(): This function is used to read a CSV file and store the data in a DataFrame.

```python
import pandas as pd

# Read a CSV file and store the data in a DataFrame
df = pd.read_csv("services.csv")
df.to_csv("data.csv")
#Display the first 5 rows of the DataFrame
print(df.head())
```

groupby(): This function is used to group the data in a DataFrame based on one or more columns, and perform aggregate functions on each group.

```python
# Group the data by the "category" column and calculate the mean of the "value" column for each group
grouped_df = df.groupby("audience")["id"].mean()

# Display the resulting DataFrame
print(grouped_df)
```

pivot_table(): This function is used to create a pivot table from a DataFrame, which shows the relationship between two or more columns.

```python
pivot_table = df.pivot_table(values="id", index="description", columns="audience", aggfunc="mean")

# Display the resulting pivot table
print(pivot_table)
```

dropna(): This function is used to remove rows or columns with missing data (NaN values) from a DataFrame.

```python
df = df.dropna()

# Display the resulting DataFrame
print(df)
```

merge(): This function is used to merge two DataFrames based on a common column.

```python
# Read two CSV files and store the data in two separate DataFrames
df1 = pd.read_csv("services.csv")
df2 = pd.read_csv("data.csv")

# Merge the two DataFrames based on the "id" column
merged_df = pd.merge(df1, df2, on="id")

# Display the resulting merged DataFrame
print(merged_df)
```

# Question 05

DataFrame.size and DataFrame.shape are both attributes of a Pandas DataFrame that provide information about its dimensions, but they represent different things.

DataFrame.size returns the total number of elements in the DataFrame, which is the product of the number of rows and the number of columns. It is equivalent to DataFrame.shape[0] * DataFrame.shape[1]. This attribute returns an integer value.

For example, if a DataFrame has 5 rows and 3 columns, DataFrame.size will return 15.

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    "col1": [1, 2, 3],
    "col2": [4, 5, 6],
    "col3": [7, 8, 9]
})

# Get the size of the DataFrame
print(df.size)
```

On the other hand, DataFrame.shape returns a tuple containing the number of rows and columns in the DataFrame. The first element of the tuple is the number of rows, and the second element is the number of columns.

For example, if a DataFrame has 5 rows and 3 columns, DataFrame.shape will return (5, 3).

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    "col1": [1, 2, 3],
    "col2": [4, 5, 6],
    "col3": [7, 8, 9]
})

# Get the shape of the DataFrame
print(df.shape)
```

In summary, DataFrame.size returns the total number of elements in the DataFrame, while DataFrame.shape returns a tuple of the number of rows and columns.

```python
import pandas as pd

def select_january_data(df):
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    return df.loc[mask]
```
