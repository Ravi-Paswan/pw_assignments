ðŸ’¡ **Q1. What is the difference between Compiler and Interpreter?**
- A compiler is a program that translates the entire source code into machine code before execution. It scans the entire source code, detects errors, and generates an executable file or binary code.
- An interpreter, on the other hand, executes the source code line by line without translating it into machine code beforehand. It interprets and executes each statement or instruction sequentially, and if an error is encountered, it halts execution at that point.

ðŸ’¡ **Q2. What is the difference between JDK, JRE, and JVM?**
- JDK (Java Development Kit): It is a software development kit that provides the necessary tools, libraries, and utilities for developing and compiling Java applications. It includes the JRE as well.
- JRE (Java Runtime Environment): It is an environment that provides the necessary runtime libraries, class libraries, and JVM (Java Virtual Machine) to execute Java applications or applets.
- JVM (Java Virtual Machine): It is a virtual machine that executes Java bytecode. It is responsible for interpreting the bytecode or dynamically compiling it into machine code for execution.

ðŸ’¡ **Q3. How many types of memory areas are allocated by JVM?**
- JVM allocates memory in several areas:
  1. Heap: It is the runtime data area where objects are allocated.
  2. Method Area: It stores class structures, method information, and constant pool.
  3. Stack: Each thread has its own stack, which stores method calls, local variables, and partial results.
  4. PC Register: It holds the address of the currently executing instruction.
  5. Native Method Stack: It contains native method information.
  6. Runtime Constant Pool: It stores constant values and references used by the code.

ðŸ’¡ **Q4. What is JIT compiler?**
- JIT (Just-In-Time) compiler is a part of the JVM that dynamically compiles bytecode into machine code during runtime. It analyzes the frequently executed parts of the bytecode and optimizes them for better performance. By compiling the code on-the-fly, it reduces the overhead of interpretation and improves the overall execution speed.

ðŸ’¡ **Q5. What are the various access specifiers in Java?**
- Java provides four access specifiers to control the accessibility of classes, methods, and variables:
  1. public: Accessible from anywhere.
  2. protected: Accessible within the same package and subclasses, even if they are in different packages.
  3. default (no specifier): Accessible within the same package only.
  4. private: Accessible within the same class only.

ðŸ’¡ **Q6. What is a compiler in Java?**
- A compiler in Java is a software tool that translates the human-readable source code written in Java programming language into machine-readable bytecode. It performs lexical analysis, syntax analysis, semantic analysis, and code generation to produce an executable file or bytecode that can be executed by the JVM.

ðŸ’¡ **Q7. Explain the types of variables in Java?**
- In Java, there are three types of variables:
  1. Local variables: These are declared and used within a method, constructor, or a block. They have no default values and must be initialized before use.
  2. Instance variables: Also known as member variables, these are declared inside a class but outside any method. Each instance of the class has its own copy of these variables.
  3. Static variables: These variables are associated with a class rather than with instances of the class. They are shared among all instances of the class.

ðŸ’¡ **Q8. What are the datatypes in Java?**
- Java provides two categories of datatypes:
  1. Primitive datatypes: These are predefined and built-in datatypes in Java. They include int, float, double, char, boolean, byte, short, and long.
  2. Reference datatypes: These datatypes are references to objects. They include classes, interfaces, arrays, and strings.

ðŸ’¡ **Q9. What are the identifiers in Java?**
- In Java, identifiers are names used to identify classes, variables, methods, labels, etc. Some rules for identifiers in Java are:
  - They must start with a letter, underscore (_), or dollar sign ($).
  - After the initial character, they can contain letters, digits, underscores, or dollar signs.
  - Java is case-sensitive, so uppercase and lowercase letters are distinct.
  - They cannot be a reserved keyword in Java.

ðŸ’¡ **Q10. Explain the architecture of JVM.**
- The architecture of JVM consists of the following components:
  1. Class Loader: Loads Java class files into the JVM.
  2. Execution Engine: Executes the bytecode by interpreting or dynamically compiling it.
  3. Garbage Collector: Manages memory by reclaiming unused objects.
  4. Runtime Data Areas: Include Heap, Method Area, Stack, PC Register, Native Method Stack, and Runtime Constant Pool.
  5. Native Method Interface: Enables interaction between Java code and native libraries or code written in other languages.
